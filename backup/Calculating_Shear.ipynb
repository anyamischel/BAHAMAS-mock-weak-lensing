```python
import numpy as np
import matplotlib.pyplot as plt
import Lensing as l
import astropy.units as u
import astropy.constants as const
from astropy.cosmology import Planck18
from astropy.cosmology import WMAP9
from scipy.fft import fft, ifft, fftfreq, fftshift, fft2, ifft2, ifftshift
```

```python
# Lens parameters
sigma_v = 1400*u.km/u.s # velocity dispersion
zlens = 0.3 # redshift of the lens

# Source parameters
zsource = 1.0 # redshift of the source
lens_x = 1*u.arcmin
lens_y = 0*u.arcsec

# Distance calculations
dL = Planck18.angular_diameter_distance(zlens) # distance from the viewer to the lens plane
dS = Planck18.angular_diameter_distance(zsource) # distance from the viewer to the source plane
dLS = Planck18.angular_diameter_distance_z1z2(zlens, zsource) # distance from the lens plane to the source

# Einstein ring radius
ee_r = l.ee_r(sigma_v, dL, dS)
ee_r.to('arcsec')
```

```python
# Making a coordinate grid to represent each point in space
n = 450
# edgex = ee_r*10
# edgey = ee_r*10

edgex = 15*u.arcmin
edgey = 15*u.arcmin

x0 = np.linspace(-edgex, edgex, n)
y0 = np.linspace(-edgey, edgey, n)

## combining to get a full cartesian coordinate grid ranging from (-1,-1) to (1,1)
x, y = np.meshgrid(x0,y0, indexing='ij')
print(x)
```

```python
rgrid1 = np.sqrt((x-lens_x)**2 + (y-lens_y)**2)
rgrid = (rgrid1.to_value('rad'))*dL
print((rgrid.max())/np.sqrt(2))

rgrid.max()/np.sqrt(2)
```

```python
# Calculate convergence for baseline source redshift (z=1.0)
dS_default = Planck18.angular_diameter_distance(1) 
dLS_default = Planck18.angular_diameter_distance_z1z2(zlens, 1)

## SIS case
Sigma = sigma_v**2/(2*const.G*rgrid)

## Gaussian Sigma case
# intensity_unlensed = l.gaussian_source_intensity_rot(beta_x = x, beta_y = y, rs=rs, source_x = source_x, source_y = source_y)
# Sigma = (intensity_unlensed.to_value(u.W/(u.arcsec)**2))*u.kg/(u.m*u.Mpc)

## Sigma crit of zsource =1.0
Sigma_crit = l.Sigma_crit(dS_default, dLS_default, dL)

convergence_normalized = (Sigma/Sigma_crit).to_value('')
# convergence_normalized[rgrid>4.1*u.Mpc] = 0

convergence = convergence_normalized * (l.Sigma_crit(dS, dLS, dL))/Sigma_crit
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(9,9))

# Convergence
convergence_plot = ax1.imshow(np.log10(convergence.T), origin = 'lower', extent = [-edgex.to_value('arcmin'), edgex.to_value('arcmin'), -edgey.to_value('arcmin'), edgey.to_value('arcmin')])
ax1.set_title('Convergence')
ax1.set_xlabel('x-coord (arcmin)')
ax1.set_ylabel('y-coord (arcmin)')
plt.colorbar(convergence_plot,fraction=0.046, pad=0.04)
convergence_plot.set_cmap('inferno')


# Shear
shear1, shear2, shear = l.calculate_shear(convergence, n, edgex.to_value('arcmin'), edgey.to_value('arcmin'), padding=2)

shear_plot = ax2.imshow(np.log(shear.T), origin = 'lower', extent = [-edgex.to_value('arcmin'), edgex.to_value('arcmin'), -edgey.to_value('arcmin'), edgey.to_value('arcmin')])
ax2.set_title('Shear')
ax2.set_xlabel('x-coord (arcmin)')
ax2.set(yticklabels=[])
ax2.set(ylabel=None)  # remove the y-axis label
ax2.tick_params(left=False)  # remove the ticks
# ax2.set_xlim(-4,4)
# ax2.set_ylim(-4,4)
l.draw_shearlines(center_x=x, center_y=y, shear1=shear1, shear2=shear2, length=1*u.arcmin, a=30, ax=ax2)

plt.colorbar(shear_plot,fraction=0.046, pad=0.04)
shear_plot.set_cmap('viridis')

plt.tight_layout()
```
